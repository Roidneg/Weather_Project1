# Question 11) What is the Mean value of each column against each Weather"

```python
data.head(2)
```

```python
# Assistant
# Assuming 'data' is a DataFrame and 'Weather' is a categorical column,
# we will select only numeric columns for the mean calculation.

# Group by 'Weather' and calculate the mean for numeric columns only
data.groupby('Weather').mean(numeric_only=True)  # Use numeric_only=True to avoid non-numeric columns
```


### Changes made:
- Added `numeric_only=True` to the `mean()` function to ensure that only numeric columns are considered for the mean calculation, preventing the TypeError.
